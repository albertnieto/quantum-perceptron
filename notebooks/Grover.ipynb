
1.Calcula el tamaño del espacio de búsqueda.
2.Obtiene el estado inicial y el operador de Grover.
3.Aplica el operador de Grover m veces al estado inicial para obtener el estado final.
4.Calcula las probabilidades de medir cada estado del sistema cuántico.
5.Realiza una medición basada en estas probabilidades.
6.Devuelve el índice del elemento medido, que corresponde a la solución buscada en el espacio de búsqueda.

```python
def grover(classical_oracle, m):
    """
    Entries :
        classical oracle -> A list of booleans or integers indicating which elements are marked (list / array).
        m -> The number of steps we want to perform
    Output :
        Elements measured at the end of the algorithm
    """
    N = len(classical_oracle)#Calcula la longitud del oráculo, el número de elementos que contiene. N representa el tamaño del espacio de búsqueda.

    """
    psi: el estado inicial del algoritmo de Grover en forma de vector (un array de amplitudes).
    G: el operador de Grover, que es una matriz que implementa una iteración del algoritmo de Grover.
    """
    psi,G = get_grover_operator(classical_oracle) # Grover inital state and operator
    """
    Aplica el operador de Grover G un total de m veces al estado inicial psi. 
    elevando G a la potencia m y luego multiplicando esta matriz potenciada por el vector psi (usando np.dot). 
    como resultado es el estado final psi después de m iteraciones del operador de Grover.
    """
    psi = np.dot(np.linalg.matrix_power(G,m),psi) # Final state
    
    """
    Calcula la probabilidad de medir cada estado. 
    calculando el valor absoluto al cuadrado de cada componente de psi. 
    La probabilidad de medir un estado particular está dada por el cuadrado del valor absoluto de la amplitud de ese estado.
    """
    p = np.array([np.abs(psi[i])**2 for i in range(N)])
    c = np.random.choice(N,1,p=p) # Mide el estado cuántico 0 o 1 , basado en la medición de probabilidad de p.
    return c[0]

```

grover_opti que presentan es una versión optimizada del algoritmo de Grover, tiene:

1.Calcula el tamaño del espacio de búsqueda (N) y el número de elementos marcados (M).
2.Calcula el ángulo theta que determina la probabilidad teórica de éxito del algoritmo.
3.Define una función para calcular la probabilidad teórica de éxito después de m iteraciones.
4.Construye una distribución de probabilidad p para medir cada estado, diferenciando entre elementos marcados y no marcados.
5.Realiza una medición basada en esta distribución de probabilidad.
6.Devuelve el índice del elemento medido, que corresponde a la solución buscada en el espacio de búsqueda.

```python
def grover_opti(classical_oracle, m):
    """
    Entries :
        classical oracle -> A list of booleans or integers indicating which elements are marked.
        m -> The number of steps we want to perform
    Output :
        Elements measured at the end of the algorithm
    """
    N = len(classical_oracle) # Number of elements
    M = sum(classical_oracle) # Number of marked elements/Suma los elementos de classical_oracle, lo que da el número de elementos "marcados" (soluciones)
    
    """
    theta es el ángulo tal que sin^2(theta) = M/N, que es la probabilidad de encontrar un elemento marcado en una iteración ideal.
    """
    theta = np.arcsin(np.sqrt(M/N))
    f = lambda t:np.sin(2*theta*t+theta)**2 # Theorical probability of success (finding a solución after 't' iterations)
    
    """
    p contendrá las probabilidades de medir cada estado en el sistema cuántico:

    Si classical_oracle[i] es True (o 1), es un elemento marcado, y su probabilidad es f(m) / M.
    Si classical_oracle[i] es False (o 0), no es un elemento marcado, y su probabilidad es (1 - f(m)) / (N - M).
    **asegura que la suma de las probabilidades sea 1
    """
    p = np.array([f(m)/M if classical_oracle[i] else (1-f(m))/(N-M)  for i in range(N)])
    """
    c, calcula un índice entre 0 y N-1 basado en la distribución de probabilidad p calculada anteriormente
    """
    c = np.random.choice(N,1,p=p) # Measurement
    return c[0]
```
