# Classical Kernels

```python
def sigmoid_kernel(x1, x2, alpha=1, c=0):
    """
    Compute the Sigmoid kernel between two vectors.

    Args:
        x1 (numpy.ndarray): First input vector.
        x2 (numpy.ndarray): Second input vector.
        alpha (float, optional): Scaling parameter. Defaults to 1.
        c (float, optional): Constant term. Defaults to 0.

    Returns:
        float: Sigmoid kernel value.
    """
    return np.tanh(alpha * np.dot(x1, x2) + c)

def laplacian_kernel(x, y, sigma=1.0):
    """
    Compute the Laplacian kernel between two vectors.

    Args:
        x (numpy.ndarray): First input vector.
        y (numpy.ndarray): Second input vector.
        sigma (float, optional): Bandwidth parameter. Defaults to 1.0.

    Returns:
        float: Laplacian kernel value.
    """
    return np.exp(-linalg.norm(x - y) / sigma)

def anova_kernel(x, y, D):
    """
    Compute the Anova kernel between two vectors.

    Args:
        x (numpy.ndarray): First input vector.
        y (numpy.ndarray): Second input vector.
        D (int): Dimensionality of the vectors.

    Returns:
        float: Anova kernel value.
    """
    return sum(np.prod(np.outer(x[:d], y[:d])) for d in range(D + 1))

def chi_squared_kernel(x, y):
    """
    Compute the Chi-squared kernel between two vectors.

    Args:
        x (numpy.ndarray): First input vector.
        y (numpy.ndarray): Second input vector.

    Returns:
        float: Chi-squared kernel value.
    """
    return np.exp(-np.sum((x - y)**2 / (x + y)))

def histogram_intersection_kernel(x, y):
    """
    Compute the Histogram Intersection kernel between two vectors.

    Args:
        x (numpy.ndarray): First input vector.
        y (numpy.ndarray): Second input vector.

    Returns:
        float: Histogram Intersection kernel value.
    """
    return np.sum(np.minimum(x, y))

def linear_kernel(x1, x2):
    """
    Compute the Linear kernel between two vectors.

    Args:
        x1 (numpy.ndarray): First input vector.
        x2 (numpy.ndarray): Second input vector.

    Returns:
        float: Linear kernel value.
    """
    return np.dot(x1, x2)

def polynomial_kernel(x, y, p=3):
    """
    Compute the Polynomial kernel between two vectors.

    Args:
        x (numpy.ndarray): First input vector.
        y (numpy.ndarray): Second input vector.
        p (int, optional): Degree of the polynomial. Defaults to 3.

    Returns:
        float: Polynomial kernel value.
    """
    return (1 + np.dot(x, y)) ** p

def gaussian_kernel(x, y, sigma=5.0):
    """
    Compute the Gaussian (RBF) kernel between two vectors.

    Args:
        x (numpy.ndarray): First input vector.
        y (numpy.ndarray): Second input vector.
        sigma (float, optional): Width of the Gaussian. Defaults to 5.0.

    Returns:
        float: Gaussian kernel value.
    """
    return np.exp(-linalg.norm(x-y)**2 / (2 * (sigma ** 2)))

def rbf_kernel(x, y, gamma=0.1):
    """
    Compute the Radial Basis Function (RBF) kernel between two vectors.

    Args:
        x (numpy.ndarray): First input vector.
        y (numpy.ndarray): Second input vector.
        gamma (float, optional): Inverse of the width of the Gaussian. Defaults to 0.1.

    Returns:
        float: RBF kernel value.
    """
    return np.exp(-gamma * linalg.norm(x - y)**2)
```

# Quantum Kernels

```python
def layer(x, params, wires, i0=0, inc=1):
    """Building block of the embedding ansatz"""
    i = i0
    for j, wire in enumerate(wires):
        qml.Hadamard(wires=[wire])
        qml.RZ(x[i % len(x)], wires=[wire])
        i += inc
        qml.RY(params[0, j], wires=[wire])

    qml.broadcast(unitary=qml.CRZ, pattern="ring", wires=wires, parameters=params[1])
```

```python
def ansatz(x, params, wires):
    """The embedding ansatz"""
    for j, layer_params in enumerate(params):
        layer(x, layer_params, wires, i0=j * len(wires))
```

```python
adjoint_ansatz = qml.adjoint(ansatz)
```

```python
def random_params(num_wires, num_layers):
    """Generate random variational parameters in the shape for the ansatz."""
    return np.random.uniform(0, 2 * np.pi, (num_layers, 2, num_wires), requires_grad=True)
```

```python
num_wires = 5
dev = qml.device("default.qubit", wires=num_wires, shots=None)
wires = dev.wires.tolist()

@qml.qnode(dev)
def kernel_circuit(x1, x2, params):
    ansatz(x1, params, wires=wires)
    adjoint_ansatz(x2, params, wires=wires)
    return qml.probs(wires=wires)
```

```python
def quantum_embedding_kernel(x1, x2, params):
    return kernel_circuit(x1, x2, params)[0]
```

# Classifiers

```python
class Perceptron(object):
    def __init__(self, n_iter=None, n_correction=0, nb_hyperplanes=None, shuffle=True, space=False, oracle=False, quantum=False, nb_ampli=10, opti=False, gamma=0.01, kernel=None, activation=step_activation):
        # -------- Perceptron definition --------
        self.weights = np.array([0.])  # Default hyperplane
        self.bias = np.array([0.])
        self.activation = activation
        self.oracle = oracle

        # -------- Space version parameters --------
        self.space = space
        self.separators = None  # Set of separators
        self.nb_hyperplanes = nb_hyperplanes
        self.selected_separator = 0  # The selected separator

        # -------- Kernel parameters --------
        self.kernel = kernel

        # -------- Training parameters --------
        self.n_iter = n_iter  # Number of iteration (include the number of steps used for the search)
        self.gamma = gamma

        # -------- Quantum parameters --------
        self.quantum = quantum
        self.nb_ampli = nb_ampli  # The amplification parameter for the quantum search
        self.opti = opti  # If we use the opti Grover

    def __str__(self):
        ptype = "Quantum" if self.quantum else "Kernel" if self.kernel else "Classical"
        stype = " Space" if self.space else " Online"
        ktype = f", kernel={self.kernel.__name__}" if self.kernel else ""
        
        return f"{ptype}{stype} perceptron{ktype}"

    def gram_matrix(self, n_samples, X, y):
        K = np.zeros((n_samples, n_samples))
        for i in range(n_samples):
            for j in range(n_samples):
                K[i,j] = self.kernel(X[i], X[j])
        return K
    
    def initialize_separators(self, X):
        len_data = len(X[0])
        self.separators = np.random.multivariate_normal([0]*len_data,np.eye(len_data),size=self.nb_hyperplanes)
        np.random.shuffle(self.separators)
        self.weights = self.separators[self.selected_separator]

    def update_support_vectors(self, X, y):
        sv = self.alpha > 1e-5
        self.alpha = self.alpha[sv]
        self.sv = X[sv]
        self.sv_y = y[sv]

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features, dtype=np.float64)
            
        K = self.gram_matrix(n_samples, X, y) if self.kernel else None

        for _ in range(self.n_iter):
            for i in range(n_samples):
                if self.predict(X[i])[0] != y[i]:
                    self.update_weights_bias(X[i], y[i]) 

        if self.kernel:
            self.update_support_vectors(X, y)

        return self.n_iter

    def update_weights_bias(self, Xi, yi):
        self.weights += yi * Xi
        self.bias += yi

    def project(self, X):
        if self.kernel:
            y_predict = np.zeros(len(X))
            for i in range(len(X)):
                s = 0
                for a, sv_y, sv in zip(self.alpha, self.sv_y, self.sv):
                    s += a * sv_y * self.kernel(X[i], sv)
                y_predict[i] = s
            return y_predict
        
        return np.dot(X, self.weights) + self.bias

    def predict(self, X):
        X = np.atleast_2d(X)
        return self.activation(self.project(X))
```

```python
class Perceptron_Albert(object):

    def __init__(self, n_iter=1):
        self.n_iter = n_iter

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.w = np.zeros(n_features, dtype=np.float64)
        self.b = 0.0

        for t in range(self.n_iter):
            for i in range(n_samples):
                if self.predict(X[i])[0] != y[i]:
                    self.w += y[i] * X[i]
                    self.b += y[i]

    def project(self, X):
        return np.dot(X, self.w) + self.b

    def predict(self, X):
        X = np.atleast_2d(X)
        return np.sign(self.project(X))
```

```python
class Kernel_Perceptron_Albert(object):

    def __init__(self, kernel, n_iter=1, params=None):
        self.kernel = kernel
        self.n_iter = n_iter
        self.params = params

    def fit(self, X, y):
        n_samples, _ = X.shape
        self.alpha = np.zeros(n_samples, dtype=np.float64)

        # Gram matrix
        K = np.zeros((n_samples, n_samples))
        kernel_func = (lambda x_, y_: self.kernel(x_, y_, self.params)) if self.params is not None else self.kernel

        for i in range(n_samples):
            for j in range(n_samples):
                    K[i,j] = kernel_func(X[i], X[j])

        for _ in range(self.n_iter):
            for i in range(n_samples):
                if np.sign(np.sum(K[:,i] * self.alpha * y)) != y[i]:
                    self.alpha[i] += 1.0

        # Support vectors
        sv = self.alpha > 1e-5
        self.alpha = self.alpha[sv]
        self.sv = X[sv]
        self.sv_y = y[sv]
        print(f"{len(self.alpha)} support vectors out of {n_samples} points")

    def project(self, X):
        y_predict = np.zeros(len(X))
        kernel_func = (lambda x_, y_: self.kernel(x_, y_, self.params)) if self.params is not None else self.kernel

        for i in range(len(X)):
            s = 0
            for a, sv_y, sv in zip(self.alpha, self.sv_y, self.sv):
                s += a * sv_y * kernel_func(X[i], sv)
            y_predict[i] = s
        return y_predict

    def predict(self, X):
        X = np.atleast_2d(X)
        return np.sign(self.project(X))
```
